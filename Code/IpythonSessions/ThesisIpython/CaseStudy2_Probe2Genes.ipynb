```python
def probes2genes(expressionFile, probe2geneFile):
    from csv import writer
    genes2probes = {}
    probes2genesDict = {}
    with open(probe2geneFile,"rb",-1) as f:
        lines = [line.strip().split("\t") for line in f ]
        for i in lines:
            temp = probes2genesDict.get(i[0].upper(), [])
            temp.append(i[1].upper())
            probes2genesDict[i[0].upper()] = list(set(temp))
            temp = genes2probes.get(i[1].upper(), [])
            temp.append(i[0].upper())
            genes2probes[i[1].upper()] = list(set(temp))
        f.close()
    
    with open(expressionFile, "rb", -1) as f:
        lines = [line.strip().split("\t") for line in f]
        with open("genes_"+expressionFile, "wb", -1) as o:
            write = writer(o, delimiter="\t")
            write.writerow(lines[0])
            for i in lines:
                try:
                    if probes2genesDict[i[0].upper()] != None:
                        for l in probes2genesDict[i[0].upper()]:
                            write.writerow( [l] + i[1:])
                except KeyError:
                    continue
            o.close()
        f.close()    
    #return probes2genesDict , genes2probes
    return
```

```python
def filterExpressionFile(geneFile, expressionFile):
    from csv import writer
    geneList = []
    with open(geneFile, "rb",-1) as f:
        geneList = [line.strip().upper() for line in f]
    with open(expressionFile,"rb",-1) as f:
        lines = [line.strip().split("\t") for line in f]
        with open("filtered"+expressionFile,"wb",-1) as o:
            write = writer(o,delimiter="\t")
            write.writerow(lines[0])
            for i in lines:
                if i[0].upper() in geneList:
                    write.writerow(i)
        o.close()
    f.close()
    return
        
```
