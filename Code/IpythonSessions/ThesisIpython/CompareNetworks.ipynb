```python
def netCompare( originalNetwork, newNetwork):
    
    OriginalDiffNet = {} 
    NewDiffNet  = {}
    intersectNet = {}
    originalNet = {}
    newNet = {}
    
    for i in originalNetwork:
        
        with open(i, "rb", -1) as f:
        
            lines = [line.strip().split("\t") for line in f]
        
            for line in lines:
                
                line[0] = line[0].upper()
                line[2] = line[2].upper()
        
                try: 
                
                    temp = originalNet[ line[0] ].get( line[2] , [] )
                    temp.append( line[1] )
                    originalNet[ line[0] ][ line[2] ] = list( set( temp ) )
            
                except KeyError:
            
                    originalNet[ line[0] ] = { line[2] : [ line[1] ]}
        
            f.close()
    
    count = 0
    for i in originalNet:
        for l in originalNet[i]:
            for e in originalNet[i][l]:
                count = count + 1
    print "OriginalNet:" , count
    
    for i in newNetwork:
    
        with open(i, "rb", -1) as f:
        
            lines = [line.strip().split("\t") for line in f]
        
            for line in lines:
                
                line[0] = line[0].upper()
                line[2] = line[2].upper()
                
                try: 
                    
                    temp = newNet[ line[2] ].get( line[0] , [] )
                    temp.append( line[1] )
                    newNet[ line[2] ][ line[0] ] = list( set( temp ) )
            
                except KeyError:
            
                    newNet[ line[2] ] = { line[0] : [ line[1] ]}
        
        f.close()
    
    count = 0
    for i in newNet:
        for l in newNet[i]:
            for e in newNet[i][l]:
                count = count + 1
    print "newNet:" , count
    
    for i in newNet:
        
        for l in newNet[i]:
            
            try: 
                
                if originalNet[i][l] != None:
                
                    try:
                    
                        intersectNet[i][l] =  newNet[i][l] + originalNet[i][l]  
            
                    except KeyError:
                    
                        intersectNet[i] = { l :   newNet[i][l] + originalNet[i][l]  }
                    
            except KeyError:
                
                try:
                    
                    if originalNet[l][i] != None :
                
                        try:
                
                            intersectNet[i]
                            intersectNet[i][l] =  newNet[i][l] + originalNet[l][i] 
            
                        except KeyError:
                    
                            intersectNet[i] = { l :   newNet[i][l] + originalNet[l][i]  }
                        
                except KeyError:
                        
                    try:
                    
                        NewDiffNet[i] 
                        NewDiffNet[i][l] = newNet[i][l]
                    
                    except KeyError:
                        
                        NewDiffNet[i] = { l : newNet[i][l] }
                        
    for i in originalNet:
        
        for l in originalNet[i]:
            
            try:
                
                if newNet[i][l] != None:
                
                    continue
                
            except KeyError:
                
                try:
                    
                    if newNet[l][i] != None:
                        
                        continue
                    
                except KeyError:
                    
                    try:
                        
                        OriginalDiffNet[i][l] = originalNet[i][l]
                        
                    except KeyError:
                        
                        OriginalDiffNet[i] = { l : originalNet[i][l]}
    count = 0
    
    for i in NewDiffNet:
    
        for l in NewDiffNet[i]:
                
            count = count + 1
    
    print "NewDiffNet:" , count

    count = 0
    
    for i in intersectNet:
    
        for l in intersectNet[i]:
        
            count = count + 1
    
    print "intersectNet:" , count
    
    count = 0
    
    for i in OriginalDiffNet:
    
        for l in OriginalDiffNet[i]:
            
            count = count + 1
    
    print "OriginalDiffNet:" , count  
    
    return OriginalDiffNet, NewDiffNet, intersectNet, newNet, originalNet
```
