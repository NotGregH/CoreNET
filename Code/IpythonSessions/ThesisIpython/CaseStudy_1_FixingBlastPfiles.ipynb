```python
araRice = {}
with open("Ara_Rice_revblast.txt","rb",-1) as f:
    lines = [line.strip().split("\t") for line in f]
    for i in lines:
        try:
            araRice[i[0].upper()]
            araRice[i[0].upper()][i[1].upper()] = 1
        except KeyError:
            araRice[i[0].upper()] = { i[1].upper() : 1 }
```

```python
from csv import writer
riceAra = {}
with open("Rice_Ara_revblast.txt", "rb", -1) as f:
    lines = [line.strip().split("\t") for line in f]
    with open("RiceAraRevBlast.txt","wb",-1) as o:
        write = writer(o,delimiter="\t")
        for line in lines:
            try:
                if araRice[line[1].upper()][line[0].upper()] == 1:
                    write.writerow([line[0].upper(),"revBlast",line[1].upper()])
                    try:
                        riceAra[line[0].upper()]
                        riceAra[line[0].upper()][line[1].upper()] = 1
                    except KeyError:
                        riceAra[line[0].upper()] = { line[1].upper() : 1 }
                else: 
                    continue
                    
            except KeyError:
                continue
        o.close()
    f.close()
        
```
