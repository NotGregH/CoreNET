```python
#### Gathering an missing Nodes and Rice Nodes 
with open("newKegg.sif", "rb",-1) as f:
    lines = [line.strip().split("\t") for line in f]
    for i in lines:
        if i[0].startswith("AT")
            noa[i[0]] = 2
            noa[i[2]] = 0 
        if i[2].startswith("AT")
            noa[i[2]] = 2
            noa[i[0]] = 0 
        
```
